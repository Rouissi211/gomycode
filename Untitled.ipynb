```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import metrics  
y=df["price"]
x=df["sqft_living"]
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.2, random_state=30)
#splitting the data with test size of 20% 
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
x=df["sqft_living"].values[:,np.newaxis]
y=df["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)      
model.fit(x_train,y_train) 
predicted=model.predict(x_train)
print("MSE", mean_squared_error(x_train,predicted))
print("R squared", metrics.r2_score(y_train,predicted))
#The positive regression coefficient tells us that there is a positive correlation between price and sqft_living. 
#the positive coefficient indicates that as the value of the price and sqft_living increases simultaneously. 
 
```

```
plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x,model.predict(x),color="k")
plt.show()
#the linear regression plot indicates that sqft_living and the price are dependant to each other.
#the coefficient of correlation is positive.
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
x=df["sqft_living"].values[:,np.newaxis]
y=df["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)      
model.fit(x_test,y_test) 
predicted=model.predict(x_test)
print("MSE", mean_squared_error(x_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
#the linear regression indicates that sqft_living and the price are 50% dependant to each other.
#the coefficient of correlation is positive 
```

```
x=df[["price","sqft_living"]]  
y=df["sqft_above"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40) 

model=LinearRegression()
model.fit(x_train,y_train) 
predicted=model.predict(x_test) 

print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
#the linear regression indicates that sqft_living, sqft_living and the price are 77.5% dependant to each other.
#the coefficient of correlation is positive.
```

```
from sklearn.preprocessing import PolynomialFeatures 
x= df[["price", "sqft_living"]]
y= df["sqft_above"].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)  
lg=LinearRegression()
poly=PolynomialFeatures(degree=3)

x_train_fit = poly.fit_transform(x_train)
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)

print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
#using the polynomial features we conclude that sqft_above, sqft_living and the price are dependant to each other with R squared equals to 76.4%.


```
